```python
pos = 0
neg = 0
neu = 0

for tweet in tweet_data:
    analysis = TextBlob(tweet.text)
    print(analysis.sentiment)
    if (analysis.sentiment.polarity) > 0:
        print('The tweet has positive sentiment')
        pos = pos + 1
    elif (analysis.sentiment.polarity) == 0:
        print('The tweet is neutral')
        neu = neu + 1
    else:
        print('The tweet has negative sentiment')
        neg = neg + 1
    
```

```python
plt.xlabel('Tags')
plt.ylabel('Polarity')
plt.pie([pos,neg,neu] , labels = ['pos','neg','neu'] , autopct = '%1.1f%%')
plt.show()
```
