```python
#Twitter API Consumer Keys 
import tweepy
import pandas as pd 
from textblob import TextBlob
import matplotlib.pyplot as plt 

#To connect as auth handler or jump server /reserve proxy server 
consumer_key = "my_consumer_key"
consumer_secret = "my_consumer_secret"

#From Proxy server we need to connect 
access_token = "my_access_token"
access_token_secret = "my_access_token_secret"


#Twitter API Authorization to connect to jump server of twitter
auth = tweepy.OAuthHandler(consumer_key,consumer_secret)

#We will be connecting from jump server to web server 
auth.set_access_token(access_token,access_token_secret)

#Now we will be conencting to the API strong server of twitter
api_connect = tweepy.API(auth)
```

```python
tweet_data = api_connect.search('farmers' , count = 10000)
```

```python
tweet_data
```

```python
for tweet in tweet_data:
    print(tweet.text)
```

```python
pos = 0
neg = 0
neu = 0

for tweet in tweet_data:
    analysis = TextBlob(tweet.text)
    print(analysis.sentiment)
    if (analysis.sentiment.polarity) > 0:
        print('The tweet has positive sentiment')
        pos = pos + 1
    elif (analysis.sentiment.polarity) == 0:
        print('The tweet is neutral')
        neu = neu + 1
    else:
        print('The tweet has negative sentiment')
        neg = neg + 1
    
```

```python
plt.pie([pos,neg,neu] , labels = ['Positive','Negative','Neutral'] , autopct = '%1.1f%%')
plt.axis('equal')

#Creating a donut chart 
circle = plt.Circle(xy = (0,0) , radius = 0.75 , facecolor = 'white')
plt.gca().add_artist(circle)
plt.show()
```
